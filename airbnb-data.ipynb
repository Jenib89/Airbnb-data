Drop some of the unwanted columns. These include `host id`, `id`, `country` and `country code` from the dataset.


```python
# List of columns to drop
columns_to_drop = ['host id', 'id', 'country', 'country code', 'license', 'lat', 'long' ]
```

```python
# Drop the unwanted columns
list_clean = list.drop(columns=columns_to_drop, errors='ignore')
```

```python
list_clean.info()
```

Checking Missing values, Duplicates and Datatypes of the columns in the dataframe and removing the duplicates


```python
## Check for missing values in the dataframe and display the count in ascending order.
missing_values = list_clean.isnull().sum().sort_values(ascending=True)
print("Missing Values Count:")
print(missing_values)
```

```python
## Check whether there are any duplicate values in the dataframe and if present remove them.

for col in list_clean.columns:
    if list_clean[col].dtype == 'object':
        list_clean[col].fillna('Unknown', inplace=True)  # For object columns, fill missing values with 'Unknown'
    else:
        list_clean[col].fillna(list_clean[col].median(), inplace=True)  # For numerical columns, fill missing values with median
```

```python
list_clean
```

```python
# Check for duplicate rows and remove them
before_duplicates = len(list_clean)
list_clean.drop_duplicates(inplace=True)
after_duplicates = len(list_clean)

# Display total number of records before and after removing duplicates
print(f"Total number of records before removing duplicates: {before_duplicates}")
print(f"Total number of records after removing duplicates: {after_duplicates}")
```

```python
## Display the total number of records in the dataframe after removing the duplicates.
list_clean
```

Renaming all column names to lowercase and replace the spaces in the column names with an underscore "_".
Remove the dollar sign and comma from the columns `price` and `service_fee`.


```python
## Rename the column.
list_clean.rename(columns={'availability 365': 'days_booked'}, inplace=True)
```

```python
## Convert all column names to lowercase and replace the spaces with an underscore "_"
list_clean.columns = list_clean.columns.str.lower().str.replace(' ', '_')
```

```python
## Remove the dollar sign and comma from the columns. If necessary, convert these two columns to the appropriate data type.

# Removing dollar sign and comma from 'price' and 'service_fee' columns
list_clean['price'] = list_clean['price'].replace({'\$': '', ',': ''}, regex=True)
list_clean['service_fee'] = list_clean['service_fee'].replace({'\$': '', ',': ''}, regex=True)

# Converting 'price' and 'service_fee' columns to numeric, coercing errors to NaN
list_clean['price'] = pd.to_numeric(list_clean['price'], errors='coerce')
list_clean['service_fee'] = pd.to_numeric(list_clean['service_fee'], errors='coerce')
```

```python
corrections = {
    'brookln': 'Brooklyn',
    'manhatan': 'Manhattan'
}
list_clean['neighbourhood_group'] = list_clean['neighbourhood_group'].replace(corrections)
list_clean['neighbourhood_group'].unique()
```

```python
## various room types 
room_type_counts = list_clean ['room_type'].value_counts()
print("Count of Various Room Types Available:")
print(room_type_counts)
```

```python
strict_cancellation = list_clean.groupby('room_type')['cancellation_policy'].value_counts().groupby(level=0).head(1)
print("Room type with strict cancellation policy:")
print(strict_cancellation)
```

```python
##prices by neighborhood group 
avg_price_neighborhood = list_clean.groupby('neighbourhood_group')['price'].mean().sort_values(ascending=False)
print("Average price per neighborhood group:")
print(avg_price_neighborhood)
```

```python
# average price per neighborhood group
plt.figure(figsize=(10, 6))
avg_price_neighborhood.plot(kind='bar', color='skyblue')
plt.title('Average Price per Neighborhood Group')
plt.xlabel('Neighborhood Group')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
max_reviews_per_group = list_clean.groupby('neighbourhood_group')['number_of_reviews'].max()
max_reviews_missing = list_clean[list_clean['neighbourhood_group'].isna()]['number_of_reviews'].max()

data = pd.DataFrame({
    'Neighbourhood Group': max_reviews_per_group.index,
    'Max Number of Reviews': max_reviews_per_group.values
})

missing_neighborhood_data = pd.DataFrame({
    'Neighbourhood Group': 'Missing Neighborhoods',
    'Max Number of Reviews': max_reviews_missing
}, index=[0])

data = pd.concat([data, missing_neighborhood_data], ignore_index=True)

display(data)

# Create a bar plot for maximum number of reviews per neighborhood group
plt.figure(figsize=(10, 6))
plt.bar(data['Neighbourhood Group'], data['Max Number of Reviews'], color='skyblue')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Max Number of Reviews')
plt.title('Maximum Number of Reviews per Neighbourhood Group')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
